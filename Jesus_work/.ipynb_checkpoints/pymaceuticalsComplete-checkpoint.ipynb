```python
#***************************************************************
# Jesus Zamora
# Pymaceuticals
# This program reads in data from two csv's then merges and  
# manipulates the data in order to present a visualtion. This program
# also create output images of PNG's and save them in the Output folder.
# Analysis provided at the bottom.
#***************************************************************


# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouseDrugFile = "data/mouse_drug_data.csv"
clinicalTrialFile = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouseDF =  pd.read_csv(mouseDrugFile)
clinicalDF = pd.read_csv(clinicalTrialFile)

# Combine the data into a single dataset
mergedData = pd.merge(clinicalDF, mouseDF, how='left', on='Mouse ID')


# Display the data table for preview
mergedData.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
meanTumorVolData= mergedData.groupby(['Drug', 'Timepoint']).mean()

# Convert to DataFrame
tumorVolDF = pd.DataFrame(meanTumorVolData['Tumor Volume (mm3)']).reset_index()

# Preview DataFrame
tumorVolDF.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
semTumorVolData= mergedData.groupby(['Drug', 'Timepoint']).sem()

# Convert to DataFrame
semTumorVolDF = pd.DataFrame(semTumorVolData['Tumor Volume (mm3)']).reset_index()


# Preview DataFrame
semTumorVolDF.head()

```

```python
# Minor Data Munging to Re-Format the Data Frames
meanPivot = tumorVolDF.pivot(index='Timepoint', columns='Drug', values='Tumor Volume (mm3)')
semPivot = semTumorVolDF.pivot(index='Timepoint', columns='Drug', values='Tumor Volume (mm3)')

# Preview that Reformatting worked
meanPivot.head()
```

```python
#Axes that will be used, stored in list for easier retrieval
xAxis = meanPivot.index
yAxis = (meanPivot['Capomulin'],meanPivot['Infubinol'], meanPivot['Ketapril'], meanPivot['Placebo'])
errBar = (semPivot['Capomulin'], semPivot['Infubinol'], semPivot['Ketapril'], semPivot['Placebo'])

# Generate the Plot (with Error Bars)
plt.errorbar(xAxis, yAxis[0], yerr=errBar[0],color='r',marker='o',markersize=5, linestyle='dashed', linewidth =.8, label='Capomulin')
plt.errorbar(xAxis, yAxis[1], yerr=errBar[1],color='b',marker='^',markersize=5, linestyle='dashed', linewidth =.8, label= 'Infubinol')
plt.errorbar(xAxis, yAxis[2], yerr=errBar[2],color='g',marker='s',markersize=5, linestyle='dashed', linewidth =.8, label= 'Ketapril')
plt.errorbar(xAxis, yAxis[3], yerr=errBar[3],color='k',marker='d',markersize=5, linestyle='dashed', linewidth =.8, label= 'Placebo')

#labels on the plot
plt.title('Tumor Response to Treatment')
plt.xlabel('Time (days)')
plt.ylabel('Tumor Volume (mm3)')
plt.grid(axis='y')
plt.legend(loc='upper left')

# Save the Figure
plt.savefig("Output/Figure_1.png")
plt.show()
```

```python
# Show the Figure
plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
meanMetSiteData = mergedData.groupby(['Drug','Timepoint']).mean()

# Convert to DataFrame
meanMetDF = pd.DataFrame(meanMetSiteData['Metastatic Sites']).reset_index()

# Preview DataFrame
meanMetDF.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
semMetSiteData = mergedData.groupby(['Drug','Timepoint']).sem()

# Convert to DataFrame
semMetDF = pd.DataFrame(semMetSiteData['Metastatic Sites']).reset_index()

# Preview DataFrame
semMetDF.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
meanMetPivot = meanMetDF.pivot(index='Timepoint', columns='Drug', values='Metastatic Sites' )
semMetPivot = semMetDF.pivot(index='Timepoint', columns='Drug', values='Metastatic Sites')


# Preview that Reformatting worked
meanMetPivot.head()
```

```python
#Axes that will be used, store on list for retrieval
xAxis = meanMetPivot.index
yAxis = (meanMetPivot['Capomulin'],meanMetPivot['Infubinol'], meanMetPivot['Ketapril'], meanMetPivot['Placebo'])
errBar = (semMetPivot['Capomulin'], semMetPivot['Infubinol'], semMetPivot['Ketapril'], semMetPivot['Placebo'])

# Generate the Plot (with Error Bars)
plt.errorbar(xAxis, yAxis[0], yerr=errBar[0],color='r',marker='o',markersize=5, linestyle='dashed', linewidth =.8, label='Capomulin')
plt.errorbar(xAxis, yAxis[1], yerr=errBar[1],color='b',marker='^',markersize=5, linestyle='dashed', linewidth =.8, label= 'Infubinol')
plt.errorbar(xAxis, yAxis[2], yerr=errBar[2],color='g',marker='s',markersize=5, linestyle='dashed', linewidth =.8, label= 'Ketapril')
plt.errorbar(xAxis, yAxis[3], yerr=errBar[3],color='k',marker='d',markersize=5, linestyle='dashed', linewidth =.8, label= 'Placebo')

# Save the Figure
#labels on the plot
plt.title('Metastatic Spread During Treatment')
plt.xlabel('Time (days)')
plt.ylabel('Tumor Volume (mm3)')
plt.grid(axis='y')
plt.legend(loc='upper left')


# Show the Figure
plt.savefig("Output/Figure_2.png")
plt.show()
```

![Metastatic Spread During Treatment](../Images/spread.png)

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
miceCountData = mergedData.groupby(['Drug', 'Timepoint']).count()

# Convert to DataFrame
miceCountDF = pd.DataFrame({'Mouse Count': miceCountData['Mouse ID']}).reset_index()

# Preview DataFrame
miceCountDF.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
miceCountPivot = miceCountDF.pivot(index='Timepoint', columns='Drug', values='Mouse Count')

# Preview the Data Frame
miceCountPivot.head()
```

```python
#Axes that will be used, stored in list for easier retrieval
xAxis = miceCountPivot.index

#yTemp is used for calculations and formatting purposes
yTemp = (miceCountPivot['Capomulin'],miceCountPivot['Infubinol'], miceCountPivot['Ketapril'], miceCountPivot['Placebo'])
yAxis = ((yTemp[0]/25)*100, (yTemp[1]/25)*100, (yTemp[2]/25)*100, (yTemp[3]/25)*100)

# Generate the Plot (Accounting for percentages)
plt.plot(xAxis, yAxis[0], color='r',marker='o',markersize=5, linestyle='dashed', linewidth =.8, label='Capomulin')
plt.plot(xAxis, yAxis[1], color='b',marker='^',markersize=5, linestyle='dashed', linewidth =.8, label= 'Infubinol')
plt.plot(xAxis, yAxis[2], color='g',marker='s',markersize=5, linestyle='dashed', linewidth =.8, label= 'Ketapril')
plt.plot(xAxis, yAxis[3], color='k',marker='d',markersize=5, linestyle='dashed', linewidth =.8, label= 'Placebo')

plt.title('Survival During Treatment')
plt.xlabel('Time (days)')
plt.ylabel('Survival Rate (%)')
plt.grid()
plt.legend(loc='lower left')




# Save the Figure
plt.savefig("Output/Figure_3.png")

# Show the Figure
plt.show()
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug

lastValue = meanPivot.iloc[-1]
firstValue = meanPivot.iloc[0] 

diff = (lastValue - firstValue)/firstValue
diffPer = diff*100

# Display the data to confirm
diffPer
```

```python
# Store all Relevant Percent Changes into a Tuple
perTuple = (diffPer[0],diffPer[2],diffPer[3],diffPer[5])


# Splice the data between passing and failing drugs
# List comprehensions to separate tuple into two lists
passDrug,failDrug = [i for i in perTuple if i >0], [j for j in perTuple if j<0 ]


# Orient widths. Add labels, tick marks, etc. 
labels = ['Capomulin','Infubinol', 'Ketapril', 'Placebo']
x = np.arange(len(labels))

fig, ax = plt.subplots()


bar1 = ax.bar(x[0], failDrug, width=1, color='g')
bar2 = ax.bar(x[1:], passDrug, width=1, color='r')

ax.set_title('Tumor Change Over 45 Day Treatment')
ax.set_ylabel('% Tumor Volume Change')
ax.set_xticks(x+.5) # The .5 is to set the tick at the end of the width of bar
ax.set_xticklabels(labels)
ax.set_ylim([-30,70])
ax.grid()

# Use functions to label the percentages of changes
# Note: function is from the internet
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2, height*.1, '%d%%' %int(height), ha='center', va='top', color='w')

# Call functions to implement the function calls
autolabel(bar1)
autolabel(bar2)


# Save the Figure
plt.savefig("Output/Figure_4")
fig.show()
```

Based on Figure 3, it also shows that Capomulin  had the highest survival rate out of all the drugs tested.

* The two other drugs, Infubinol and Ketapril, seemed to have very litte effect on tumor volume. In fact, the tumors started to grow under both of the medicines. In terms of ranking effectiveness against metastatic spread, it would appear that Infubonil is slightly better than Ketapril. Infobunil does end up having the worth survival rate out of all the methods tested. 

* The most surprising part is that the Placebo treatment seems to have the relatively the same effect as Infubinol and Ketapril. It followed the general trend as both treatments throughout all the figures. The Placebo actually ends up having a higher survival rate than Infubinol, as show in Figure 3. It should be noted that the metastatic spread was the highest under the Placebo but not too far from Ketrapril.
