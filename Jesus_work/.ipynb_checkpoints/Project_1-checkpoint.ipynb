```python
import pandas as pd
import numpy as np
```

```python
caFile = 'Data/CAvideos.csv'
usFile = 'Data/USvideos.csv'
mxFile = 'Data/MXvideos.csv'
inFile = 'Data/INVideos.csv'
jpFile = 'Data/JPVideos.csv'
krFile = 'Data/KRVideos.csv'


caDF = pd.read_csv(caFile, encoding='ANSI')
usDF = pd.read_csv(usFile, encoding='ANSI')
mxDF = pd.read_csv(mxFile, encoding='ANSI')
inDF = pd.read_csv(inFile, encoding='ANSI')
jpDF = pd.read_csv(jpFile, encoding='ANSI')
krDF = pd.read_csv(krFile, encoding='ANSI')

caDF.head()
```

```python
# Keeping columns that only pertain to my question, cleaning data

caCleanDF = caDF.loc[:,['video_id', 'views', 'likes', 'dislikes', 'comment_count', 'comments_disabled', 'ratings_disabled', 
                        'video_error_or_removed']]
usCleanDF = usDF.loc[:,['video_id', 'views', 'likes', 'dislikes', 'comment_count', 'comments_disabled', 'ratings_disabled', 
                        'video_error_or_removed']]
mxCleanDF = mxDF.loc[:,['video_id', 'views', 'likes', 'dislikes', 'comment_count', 'comments_disabled', 'ratings_disabled', 
                        'video_error_or_removed']]
inCleanDF = inDF.loc[:,['video_id', 'views', 'likes', 'dislikes', 'comment_count', 'comments_disabled', 'ratings_disabled', 
                        'video_error_or_removed']]
jpCleanDF = jpDF.loc[:,['video_id', 'views', 'likes', 'dislikes', 'comment_count', 'comments_disabled', 'ratings_disabled', 
                        'video_error_or_removed']]
krCleanDF = krDF.loc[:,['video_id', 'views', 'likes', 'dislikes', 'comment_count', 'comments_disabled', 'ratings_disabled', 
                        'video_error_or_removed']]

krCleanDF.head()
```
