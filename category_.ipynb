```python
#import depedencies
import pandas as pd
import csv
import numpy as np
import json
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
#read the data  asia continent using pandas
india_df=pd.read_csv('INvideos.csv',encoding='ANSI')
japan_df=pd.read_csv('JPvideos.csv',encoding='ANSI')
korea_df=pd.read_csv('KRvideos.csv',encoding='ANSI')
```

```python
#read the north america data files using pandas 
canada_df=pd.read_csv('CAvideos.csv',encoding='ANSI')
us_df=pd.read_csv('USvideos.csv',encoding='ANSI')
mexico_df=pd.read_csv('MXvideos.csv',encoding='ANSI')
```

```python
#cleaning up us data
clean_us_data=us_df.loc[:, ["video_id","category_id"]] 
clean_us_data.insert(0,'Country','United States')
pd.DataFrame(clean_us_data.rename(columns={'category_id':'us_category_id'}))
category_id_df=pd.DataFrame(clean_us_data.category_id.value_counts())
category_id_df.rename(columns={'category_id':'total_videos'})
#read us_video category json file
us_videos_categories = pd.read_json('US_category_id.json')
us_videos_categories
categories = {int(category['id']): category['snippet']['title'] for category in us_videos_categories['items']}
categories
#plot us graph 
plt.title=["United States Video Categories"]
sns.set(font_scale=1.2,rc={'figure.figsize':(15.7,8.27)})
sns_ax = sns.countplot([categories[i] for i in clean_us_data.category_id])
_, labels = plt.xticks()
_ = sns_ax.set_xticklabels(labels, rotation=60)
#save figure
fig = sns_ax.get_figure()
fig.savefig("us.png")
plt.show()


```
